## 문제 1

1) 한 글자 제외

```python
poliRdd = spark.sparkContext\
    .textFile(os.path.join("data", "20191021_policeAddress.txt"))
rm1_poliRdd = poliRdd.flatMap(lambda x:x.split())\
    .flatMap(lambda x:x.split())\
    .filter(lambda x: len(x) > 1)

for word in rm1_poliRdd.collect():
    print word
```

2) 빈도순 출력

```python
sortedList = rm1_poliRdd\
    .map(lambda x:(x,1))\
    .reduceByKey(lambda x,y:x+y)\
    .map(lambda x:(x[1],x[0]))\
    .sortByKey(False)\
    .collect()

for i in sortedList:
    print i[0],i[1]
```

## 문제 3

1) 5행 출력

```python
df = spark.read.format('com.databricks.spark.csv')\
    .options(header='true', inferschema='true').load('data/seoulBicycleDailyCount_2018_201903.csv')
df.show(5)
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

df=df\
    .withColumn("year",df.date.substr(1, 4))\
    .withColumn("month",df.date.substr(6, 2))\
    .withColumn("day",df.date.substr(9, 2))
# df.printSchema()
df.show()
```

2) 년별 분기별 대여건수 합계

```python
df.groupBy('year')\
    .pivot('month')\
    .agg({" count":"sum"}).show()

# df.groupBy('year')\
#     .agg({"Count":"sum"}).show()
```

2)년별 분기별 대여건수 합계(정답)

```python
import pyspark.sql.functions as F
df=df\
    .withColumn("Month",F.month('date'))
```

```python
# 년별 분기별 대여건수 합계 정답
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
def classifyQuarter(s):
    q=""
    if 1<=s and s< 4:
        q="Q1"
    elif 4<=s and s<7:
        q="Q2"
    elif 7<=s and s<10:
        q="Q3"
    elif 10<=s and s<=12:
        q="Q4"
    else:
        q="no"
    return q

quarter_udf = udf(classifyQuarter, StringType())
```

```python
df=df.withColumn('quarter', quarter_udf(df.Month))
# df.printSchema()
df.groupBy('year').pivot('quarter').agg({" count":"sum"}).show()
```
